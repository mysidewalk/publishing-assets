```python
import re
from operator import itemgetter

import pandas
```

```python
dataframe = pandas.read_csv('omaha-council-bluffs-msa/pivoted-characteristics.csv', index_col='shid')
cleaned_dataframe = dataframe.dropna(axis=1, how='all')
cleaned_dataframe
```

```python
correlated_df = cleaned_dataframe.corr(method='pearson', min_periods=150)
cleaned_correlated_df = correlated_df.dropna(axis=0, how='all').dropna(axis=1, how='all')
cleaned_correlated_df
```

```python
cleaned_correlated_df.to_csv('omaha-council-bluffs-msa/correlations.csv')
```
